```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)

X = np.abs(np.random.normal(100, 100, size=400))
noise = np.random.normal(0, 50, size=400)

y_data = lambda X, noise: (2*X) + 25 + noise
y = y_data(X=X, noise=noise)

data = {"called minutes per month": X, "SEK per month": y}

mask = (X < 300) & (y > 0)
X = X[mask]
y = y[mask]

df = pd.DataFrame(data=data)

sns.scatterplot(data=df, x="called minutes per month", y="SEK per month")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
