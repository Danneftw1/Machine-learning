```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('ReLU Activation Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

# Generate input values
x = np.linspace(-5, 5, 100)

# Calculate output values for sigmoid function
y_sigmoid = sigmoid(x)

# Calculate output values for ReLU function
y_relu = relu(x)

# Plot the activation functions
fig, ax = plt.subplots()
ax.plot(x, y_sigmoid, label='Sigmoid')
ax.plot(x, y_relu, label='ReLU')

# Add axis labels and a title
ax.set_xlabel('Input')
ax.set_ylabel('Output')
ax.set_title('Activation Functions')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
