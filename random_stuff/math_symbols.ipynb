# 3D Gradient Descent - ChatGPT

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function to optimize
def f(x, y):
    return (x ** 2 + y ** 2) * np.sin(x)

# Define the gradient of the function
def grad_f(x, y):
    grad_x = (2 * x + y ** 2) * np.sin(x) + (x ** 2 + y ** 2) * np.cos(x)
    grad_y = 2 * y * np.sin(x)
    return np.array([grad_x, grad_y])

# Define the learning rate
learning_rate = 0.05

# Define the starting point
start_point = np.array([2, 2])

# Perform gradient descent
num_iterations = 30
trajectory = [start_point]
for i in range(num_iterations):
    grad = grad_f(*trajectory[-1])
    new_point = trajectory[-1] - learning_rate * grad
    trajectory.append(new_point)

# Plot the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
ax.plot_surface(X, Y, Z, cmap='viridis')

# Plot the trajectory of the gradient descent
trajectory = np.array(trajectory)
ax.plot(trajectory[:, 0], trajectory[:, 1], f(*trajectory.T), color='r', marker='o')

plt.show()
```
