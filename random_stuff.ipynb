```python
import pygame
import math

# Define the dimensions of the window
WINDOW_WIDTH = 400
WINDOW_HEIGHT = 400

# Initialize Pygame
pygame.init()

# Create the window
window = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("Spinning Donut")

# Define the colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
BROWN = (102, 14, 30)

# Define the parameters of the donut
radius1 = 100
radius2 = 50
k2 = 7
k1 = WINDOW_HEIGHT // 2

# Define the main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the window
    window.fill(BLACK)

    # Create the donut
    for theta in range(0, 360, 5):
        for phi in range(0, 360, 5):
            x = math.cos(math.radians(theta)) * (radius1 + math.cos(math.radians(phi)) * radius2) + WINDOW_WIDTH // 2
            y = math.sin(math.radians(theta)) * (radius1 + math.cos(math.radians(phi)) * radius2) + k1
            z = k2 + math.sin(math.radians(phi)) * radius2
            size = 0.05 * z
            pygame.draw.circle(window, BROWN, (int(x), int(y)), max(1, int(size)))

    # Spin the donut
    radius1 -= 1

    # Update the display
    pygame.display.update()

# Quit Pygame
pygame.quit()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to optimize
def f(x):
    return x ** 2 + 5 * np.sin(x)

# Define the derivative of the function
def df(x):
    return 2 * x + 5 * np.cos(x)

# Define the gradient descent function
def gradient_descent(f, df, x, alpha, num_iterations):
    history = [x]
    for i in range(num_iterations):
        x = x - alpha * df(x)
        history.append(x)
    return history

# Set the initial value of x and the learning rate
x_init = -5
learning_rate = 0.2

# Run gradient descent for 50 iterations
history = gradient_descent(f, df, x_init, learning_rate, 50)

# Plot the function and the gradient descent history
x = np.linspace(-5, 5, 100)
plt.plot(x, f(x), 'b', label='f(x)')
plt.plot(history, f(np.array(history)), 'r+', markersize=12, label='Gradient descent')
plt.legend()
plt.show()
```
